# Evaluación de modelos de regresión: Error Absoluto Medio vs Error Cuadrático Medio

Los modelos de regresión a menudo se evalúan utilizando métricas como el Error Absoluto Medio (MAE) y el Error Cuadrático Medio (MSE). En esta lección, ilustraremos la diferencia entre estas dos métricas y demostraremos cómo MSE castiga más los errores grandes en comparación con MAE.

Primero, generaremos un conjunto de datos sintético y ajustaremos un modelo de regresión lineal a este conjunto de datos.


```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Crear conjunto de datos sintético
np.random.seed(0)
X = np.random.rand(10, 1) * 10
y = 3 * X + 4 + np.random.randn(10, 1) * 3

# Ajustar la regresión lineal
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
```

## Visualización de Error Absoluto Medio (MAE)

El Error Absoluto Medio (MAE) se calcula como el promedio de los errores absolutos. En la siguiente gráfica, los errores absolutos se representan como líneas verdes que se extienden desde los puntos de datos reales hasta la línea de regresión.


```python
fig, axs = plt.subplots(1, 2, figsize=(15, 7))

# Gráfica 1: Error absoluto
axs[0].scatter(X, y, color='blue')
axs[0].plot(X, y_pred, color='red')
for i in range(X.shape[0]):
    axs[0].plot([X[i, 0], X[i, 0]], [y[i, 0], y_pred[i, 0]], color='green')

axs[0].set_title('Error Absoluto (MAE)')
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')

# Gráfica 2: Error cuadrático
axs[1].scatter(X, y, color='blue')
axs[1].plot(X, y_pred, color='red')
for i in range(X.shape[0]):
    error = y[i, 0] - y_pred[i, 0]
    # Decide la dirección de la línea en función de si el valor real está por encima o por debajo de la predicción
    direction = -1 if y[i, 0] < y_pred[i, 0] else 1
    axs[1].plot([X[i, 0], X[i, 0]], [y_pred[i, 0], y_pred[i, 0] + direction * error**2], color='purple')

axs[1].set_title('Error Cuadrático (MSE)')
axs[1].set_xlabel('X')
axs[1].set_ylabel('Y')

plt.show()
```
